````python
from typing import Optional, Literal
from pydantic import (
    Field,
    SecretStr,
    model_validator,
)
from pydantic_settings import (
    BaseSettings,
    SettingsConfigDict,
)


class EnterpriseSettings(BaseSettings):

    # =====================================================
    # Application
    # =====================================================

    APP_NAME: str = "Enterprise RAG Platform"

    APP_VERSION: str = "1.0.0"

    ENVIRONMENT: Literal[
        "development",
        "staging",
        "production"
    ] = "development"

    DEBUG: bool = False

    # =====================================================
    # Multi Tenant
    # =====================================================

    MULTI_TENANT_ENABLED: bool = True

    TENANT_HEADER: str = "X-Tenant-ID"

    DEFAULT_TENANT: str = "default"

    # =====================================================
    #  LLM Providers
    # =====================================================
    LLM_PROVIDER: Literal[
        "openai",
        "azure_openai",
        "anthropic",
        "ollama",
        "vllm",
    ] =   "ollama"

    LLM_MODEL: str = "qwen2.5-coder:7b"

    LLM_BASE_URL: str = "http://localhost:11434"

    TEMPERATURE: float = 0.0

    MAX_GENERATION_TOKENS: int = 4096

    MAX_CONTEXT_WINDOW: int = 32000


    # =====================================================
    # Embeddings
    # =====================================================

    EMBEDDING_PROVIDER: Literal[
     "openai",
     "jina",
     "local"
  ] = "local"
    LOCAL_EMBEDDING_MODEL: str = (
    "BAAI/bge-small-en-v1.5"
  )
    EMBEDDING_MODEL: str = "text-embedding-3-large"

    EMBEDDING_DIMENSIONS: Optional[int] = None

    EMBEDDING_BATCH_SIZE: int = 32

    EMBEDDING_VERSION: str = "v1"


    OPENAI_API_KEY: Optional[SecretStr] = None

    JINA_API_KEY: Optional[SecretStr] = None

    # =====================================================
    # Qdrant
    # =====================================================

    QDRANT_URL: str

    QDRANT_API_KEY: SecretStr

    QDRANT_COLLECTION: str = "documents"

    # =====================================================
    # OpenSearch
    # =====================================================

    OPENSEARCH_HOST: str

    OPENSEARCH_PORT: int = Field(
        default=9200,
        ge=1,
        le=65535,
    )

    OPENSEARCH_USER: Optional[str] = None

    OPENSEARCH_PASSWORD: Optional[
        SecretStr
    ] = None

    INDEX_NAME: str = "enterprise_docs"

    # =====================================================
    # Redis
    # =====================================================

    REDIS_URL: str = (
        "redis://localhost:6379/0"
    )

    CACHE_TTL_SECONDS: int = 3600

    # =====================================================
    # Retrieval
    # =====================================================

    RETRIEVAL_TOP_K: int = 50

    RERANK_TOP_K: int = 20

    FINAL_CONTEXT_DOCS: int = 5

    # =====================================================
    # Chunking
    # =====================================================

    CHUNK_SIZE: int = 400

    CHUNK_OVERLAP: int = 50

    PARENT_CHUNK_SIZE: int = 2000

    PARENT_CHUNK_OVERLAP: int = 200

    # =====================================================
    # Security
    # =====================================================

    DEFAULT_SECURITY_LEVEL: Literal[
        "public",
        "internal",
        "confidential",
        "restricted",
    ] = "internal"

    ENCRYPT_DOCUMENTS_AT_REST: bool = True

    ENCRYPT_CACHE_DATA: bool = True

    AUDIT_LOGGING_ENABLED: bool = True

    AUDIT_RETENTION_DAYS: int = 365

    JWT_SECRET: Optional[SecretStr] = None

    JWT_ALGORITHM: str = "HS256"

    SESSION_TIMEOUT_MINUTES: int = 60

    # =====================================================
    # Auth0 / OAuth
    # =====================================================

    AUTH_PROVIDER: Literal[
        "auth0",
        "azure_ad",
        "okta",
    ] = "auth0"

    AUTH0_DOMAIN: Optional[str] = None

    AUTH0_AUDIENCE: Optional[str] = None

    AUTH0_CLIENT_ID: Optional[str] = None

    AUTH0_CLIENT_SECRET: Optional[
        SecretStr
    ] = None

    # =====================================================
    # Rate Limiting
    # =====================================================

    RATE_LIMIT_ENABLED: bool = True

    REQUESTS_PER_MINUTE: int = 60

    REQUESTS_PER_HOUR: int = 1000

    # =====================================================
    # Cost Controls
    # =====================================================

    DAILY_TOKEN_BUDGET: int = 10_000_000

    MONTHLY_COST_LIMIT_USD: float = 1000.0

    # =====================================================
    # Storage (AWS S3)
    # =====================================================

    AWS_REGION: str = "us-east-1"

    S3_BUCKET: Optional[str] = None

    AWS_ACCESS_KEY_ID: Optional[
        SecretStr
    ] = None

    AWS_SECRET_ACCESS_KEY: Optional[
        SecretStr
    ] = None

    # =====================================================
    # Azure Storage
    # =====================================================

    AZURE_STORAGE_ACCOUNT: Optional[
        str
    ] = None

    AZURE_STORAGE_KEY: Optional[
        SecretStr
    ] = None

    # =====================================================
    # Kafka
    # =====================================================

    KAFKA_ENABLED: bool = False

    KAFKA_BOOTSTRAP_SERVERS: Optional[
        str
    ] = None

    KAFKA_DOCUMENT_TOPIC: str = (
        "documents"
    )

    # =====================================================
    # Observability
    # =====================================================

    ENABLE_TRACING: bool = True

    ENABLE_METRICS: bool = True

    ENABLE_PROMETHEUS: bool = True

    ENABLE_LANGSMITH: bool = True

    LANGSMITH_API_KEY: Optional[
        SecretStr
    ] = None

    ENABLE_OPENTELEMETRY: bool = True

    ENABLE_GRAFANA: bool = True

    TRACE_SAMPLE_RATE: float = 1.0

    # =====================================================
    # Feature Flags
    # =====================================================

    ENABLE_HYBRID_SEARCH: bool = True

    ENABLE_RERANKING: bool = True

    ENABLE_QUERY_EXPANSION: bool = True

    ENABLE_SELF_QUERY: bool = True

    ENABLE_CACHING: bool = True

    ENABLE_CITATIONS: bool = True

    ENABLE_AGENTIC_RAG: bool = False

    ENABLE_GRAPH_RETRIEVAL: bool = False

    ENABLE_NEW_RERANKER: bool = False

    # =====================================================
    # Disaster Recovery
    # =====================================================

    BACKUP_ENABLED: bool = True

    BACKUP_INTERVAL_HOURS: int = 24

    BACKUP_RETENTION_DAYS: int = 30

    # =====================================================
    # Workers
    # =====================================================

    GPU_WORKERS: int = 1

    CPU_WORKERS: int = 4

    REQUEST_TIMEOUT: int = 60

    MAX_RETRIES: int = 3

    # =====================================================
    # Pydantic Config
    # =====================================================

    model_config = SettingsConfigDict(
        env_file=".env",
        env_file_encoding="utf-8",
        extra="ignore",
        case_sensitive=False,
    )

    # =====================================================
    # Validators
    # =====================================================

    @model_validator(mode="after")
    def validate_embedding_provider(self):
        if (
            self.EMBEDDING_PROVIDER == "openai"
            and self.OPENAI_API_KEY is None
        ):
            raise ValueError(
                "OPENAI_API_KEY is required"
            )

        if (
            self.EMBEDDING_PROVIDER == "jina"
            and self.JINA_API_KEY is None
        ):
            raise ValueError(
                "JINA_API_KEY is required"
            )

        return self

    @model_validator(mode="after")
    def validate_chunking(self):

        if self.CHUNK_OVERLAP >= self.CHUNK_SIZE:
            raise ValueError(
                "Invalid chunk configuration"
            )

        if (
            self.PARENT_CHUNK_OVERLAP
            >= self.PARENT_CHUNK_SIZE
        ):
            raise ValueError(
                "Invalid parent chunk configuration"
            )

        return self

    @model_validator(mode="after")
    def validate_retrieval(self):

        if (
            self.RERANK_TOP_K
            > self.RETRIEVAL_TOP_K
        ):
            raise ValueError(
                "RERANK_TOP_K > RETRIEVAL_TOP_K"
            )

        return self

    @property
    def embedding_dimensions(self):

        if self.EMBEDDING_DIMENSIONS:
            return self.EMBEDDING_DIMENSIONS

        return {
            "text-embedding-3-small": 1536,
            "text-embedding-3-large": 3072,
        }.get(
            self.EMBEDDING_MODEL,
            3072,
        )

from functools import lru_cache

@lru_cache
def get_settings():
    return EnterpriseSettings()
````

````python
"""
Enterprise Logging System - Production Ready
"""

import json
import logging
import re
import sys
import asyncio

from time import perf_counter
from pathlib import Path
from functools import wraps
from contextvars import ContextVar
from logging.handlers import RotatingFileHandler

# ==========================================================
# Context Variables
# ==========================================================

request_id_ctx = ContextVar("request_id", default=None)
tenant_id_ctx = ContextVar("tenant_id", default=None)
user_id_ctx = ContextVar("user_id", default=None)
trace_id_ctx = ContextVar("trace_id", default=None)

# ==========================================================
# Context Helpers
# ==========================================================

def set_request_context(
    request_id: str | None = None,
    tenant_id: str | None = None,
    user_id: str | None = None,
    trace_id: str | None = None,
):
    request_id_ctx.set(request_id)
    tenant_id_ctx.set(tenant_id)
    user_id_ctx.set(user_id)
    trace_id_ctx.set(trace_id)

# ==========================================================
# Context Filter
# ==========================================================

class ContextFilter(logging.Filter):
    def filter(self, record):
        record.request_id = request_id_ctx.get()
        record.tenant_id = tenant_id_ctx.get()
        record.user_id = user_id_ctx.get()
        record.trace_id = trace_id_ctx.get()
        return True

# ==========================================================
# PII Redaction Patterns
# ==========================================================

class PIIUtils:
    EMAIL_PATTERN = re.compile(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}")
    API_KEY_PATTERN = re.compile(r"(sk-[A-Za-z0-9]{20,}|AIzaSy[A-Za-z0-9-_]{33})")
    CREDIT_CARD_PATTERN = re.compile(r"\b(?:\d[ -]*?){13,16}\b")

    @classmethod
    def redact_text(cls, text: str) -> str:
        if not text:
            return text
        text = cls.EMAIL_PATTERN.sub("[REDACTED_EMAIL]", text)
        text = cls.API_KEY_PATTERN.sub("[REDACTED_API_KEY]", text)
        text = cls.CREDIT_CARD_PATTERN.sub("[REDACTED_CARD]", text)
        return text

class PIIRedactionFilter(logging.Filter):
    def filter(self, record):
        # جلب الرسالة مدمجاً معها الـ args إن وجدت
        message = str(record.getMessage())
        # حجب البيانات
        record.msg = PIIUtils.redact_text(message)
        # تفريغ الـ args تماماً لمنع الفورمات المزدوج أو الأخطاء البرمجية
        record.args = ()
        return True

# ==========================================================
# Safe JSON Encoder
# ==========================================================

class SafeJsonEncoder(json.JSONEncoder):
    """يمنع انهيار النظام إذا تم تمرير كائنات معقدة في الـ extra fields"""
    def default(self, obj):
        try:
            if isinstance(obj, Path):
                return str(obj)
            return super().default(obj)
        except TypeError:
            return str(obj)

# ==========================================================
# JSON Formatter
# ==========================================================

class JsonFormatter(logging.Formatter):
    def format(self, record):
        payload = {
            "timestamp": self.formatTime(record),
            "level": record.levelname,
            "logger": record.name,
            "message": record.getMessage(),
            "module": record.module,
            "function": record.funcName,
            "line": record.lineno,
            "request_id": getattr(record, "request_id", None),
            "tenant_id": getattr(record, "tenant_id", None),
            "user_id": getattr(record, "user_id", None),
            "trace_id": getattr(record, "trace_id", None),
        }

        # تجميع الـ extra fields تلقائياً باستثناء الحقول الأساسية لبايثون
        extra_fields = {}
        for key, value in record.__dict__.items():
            if key.startswith("_"):
                continue
            if key in {
                "args", "msg", "message", "pathname", "filename", "module",
                "exc_info", "exc_text", "stack_info", "lineno", "funcName",
                "created", "msecs", "relativeCreated", "thread", "threadName",
                "processName", "process", "name", "levelname", "levelno",
                "request_id", "tenant_id", "user_id", "trace_id"
            }:
                continue
            extra_fields[key] = value

        if extra_fields:
            payload["extra"] = extra_fields

        # التعامل الآمن وحجب الـ PII من الـ Exceptions والـ Tracebacks
        if record.exc_info:
            exc_text = self.formatException(record.exc_info)
            payload["exception"] = PIIUtils.redact_text(exc_text)

        return json.dumps(payload, cls=SafeJsonEncoder, ensure_ascii=False)

# ==========================================================
# Logging Setup
# ==========================================================

def setup_logging(log_level="INFO", log_file=None):
    logger = logging.getLogger("enterprise-rag")
    logger.setLevel(getattr(logging, log_level.upper(), logging.INFO))
    logger.propagate = False

    if logger.handlers:
        return logger

    context_filter = ContextFilter()
    pii_filter = PIIRedactionFilter()

    # Console Handler
    console_handler = logging.StreamHandler(sys.stdout)
    console_handler.setFormatter(JsonFormatter())
    console_handler.addFilter(context_filter)
    console_handler.addFilter(pii_filter)
    logger.addHandler(console_handler)

    # File Handler
    if log_file:
        Path(log_file).parent.mkdir(parents=True, exist_ok=True)
        file_handler = RotatingFileHandler(
            log_file,
            maxBytes=50 * 1024 * 1024, # 50MB
            backupCount=10,
            encoding="utf-8",
        )
        file_handler.setFormatter(JsonFormatter())
        file_handler.addFilter(context_filter)
        file_handler.addFilter(pii_filter)
        logger.addHandler(file_handler)

    return logger

# ==========================================================
# Performance Decorator (Sync / Async Smart Support)
# ==========================================================

def log_execution_time(logger):
    def decorator(func):
        if asyncio.iscoroutinefunction(func):
            @wraps(func)
            async def async_wrapper(*args, **kwargs):
                start = perf_counter()
                try:
                    return await func(*args, **kwargs)
                finally:
                    elapsed = (perf_counter() - start) * 1000
                    logger.info(
                        f"{func.__name__}_completed",
                        extra={"latency_ms": round(elapsed, 2)},
                    )
            return async_wrapper
        else:
            @wraps(func)
            def sync_wrapper(*args, **kwargs):
                start = perf_counter()
                try:
                    return func(*args, **kwargs)
                finally:
                    elapsed = (perf_counter() - start) * 1000
                    logger.info(
                        f"{func.__name__}_completed",
                        extra={"latency_ms": round(elapsed, 2)},
                    )
            return sync_wrapper
    return decorator

# ==========================================================
# Metrics Helpers
# ==========================================================

def log_retrieval_metrics(logger, query, docs_count, latency_ms, reranked):
    logger.info(
        "retrieval_complete",
        extra={
            "query": PIIUtils.redact_text(query),  # تأمين الاستعلامات
            "docs_count": docs_count,
            "latency_ms": latency_ms,
            "reranked": reranked,
        },
    )

def log_embedding_metrics(logger, chunks_count, latency_ms):
    logger.info(
        "embedding_complete",
        extra={
            "chunks_count": chunks_count,
            "latency_ms": latency_ms,
        },
    )

# Global Initialization
logger = setup_logging(log_level="INFO", log_file="logs/app.log")
````

````python
from dataclasses import dataclass, field
from typing import Dict, Any, Optional, List, Literal
from uuid import uuid4
from datetime import datetime, timezone
import hashlib


# ==========================================================
# Enterprise Document
# ==========================================================

@dataclass
class EnterpriseDocument:

    id: str = field(
        default_factory=lambda: str(uuid4())
    )

    content: str = ""

    source: str = ""

    metadata: Dict[str, Any] = field(
        default_factory=dict
    )

    # ----------------------------------
    # Multi Tenant
    # ----------------------------------

    tenant_id: str = "default"

    owner_id: Optional[str] = None

    # ----------------------------------
    # Security
    # ----------------------------------

    security_level: Literal[
        "public",
        "internal",
        "confidential",
        "restricted"
    ] = "internal"

    allowed_roles: List[str] = field(
        default_factory=lambda: ["viewer"]
    )

    # ----------------------------------
    # Document Metadata
    # ----------------------------------

    document_type: str = "text"

    language: str = "en"

    tags: List[str] = field(
        default_factory=list
    )

    # ----------------------------------
    # Versioning
    # ----------------------------------

    version: int = 1

    embedding_version: str = "v1"

    # ----------------------------------
    # Lifecycle
    # ----------------------------------

    status: str = "active"

    created_at: datetime = field(
        default_factory=lambda:
        datetime.now(timezone.utc)
    )

    updated_at: datetime = field(
        default_factory=lambda:
        datetime.now(timezone.utc)
    )

    # ----------------------------------
    # Integrity
    # ----------------------------------

    content_hash: Optional[str] = None

    checksum_algorithm: str = "sha256"

    # ----------------------------------
    # Auto Initialization
    # ----------------------------------

    def __post_init__(self):

        if self.content:
            self.generate_hash()

    # ----------------------------------
    # Utility Methods
    # ----------------------------------

    def generate_hash(self) -> str:

        self.content_hash = hashlib.sha256(
            self.content.encode("utf-8")
        ).hexdigest()

        return self.content_hash

    def to_metadata(self) -> Dict[str, Any]:

        system_meta = {
            "document_id": self.id,
            "tenant_id": self.tenant_id,
            "owner_id": self.owner_id,
            "security_level": self.security_level,
            "version": self.version,
            "embedding_version": self.embedding_version,
            "document_type": self.document_type,
            "language": self.language,
            "tags": self.tags,
            "source": self.source,
            "created_at": self.created_at.isoformat(),
            "updated_at": self.updated_at.isoformat(),
        }

        return {
            **self.metadata,
            **system_meta,
        }


# ==========================================================
# Document Chunk
# ==========================================================

@dataclass
class DocumentChunk:

    id: str = field(
        default_factory=lambda: str(uuid4())
    )

    document_id: str = ""

    tenant_id: str = "default"

    content: str = ""

    chunk_index: int = 0

    metadata: Dict[str, Any] = field(
        default_factory=dict
    )

    # ----------------------------------
    # Parent Retrieval Strategy
    # ----------------------------------

    parent_chunk_id: Optional[str] = None

    # ----------------------------------
    # Embeddings
    # ----------------------------------

    embedding: Optional[List[float]] = None

    embedding_model: Optional[str] = None

    embedding_version: str = "v1"

    # ----------------------------------
    # Context Window
    # ----------------------------------

    token_count: Optional[int] = None

    # ----------------------------------
    # Security
    # ----------------------------------

    security_level: Literal[
        "public",
        "internal",
        "confidential",
        "restricted"
    ] = "internal"

    # ----------------------------------
    # Audit
    # ----------------------------------

    created_at: datetime = field(
        default_factory=lambda:
        datetime.now(timezone.utc)
    )

    def to_dict(self) -> Dict[str, Any]:

        return {
            "id": self.id,
            "document_id": self.document_id,
            "tenant_id": self.tenant_id,
            "content": self.content,
            "chunk_index": self.chunk_index,
            "metadata": self.metadata,
            "parent_chunk_id": self.parent_chunk_id,
            "embedding_version": self.embedding_version,
            "security_level": self.security_level,
            "token_count": self.token_count,
            "created_at": self.created_at.isoformat(),
        }


# ==========================================================
# Retrieval Result
# ==========================================================

@dataclass
class RetrievalResult:

    chunk: DocumentChunk

    vector_score: float = 0.0

    bm25_score: float = 0.0

    rerank_score: Optional[float] = None

    final_score: float = 0.0

    retrieval_source: str = "hybrid"

    retrieved_at: datetime = field(
        default_factory=lambda:
        datetime.now(timezone.utc)
    )
````

````python
from __future__ import annotations

from abc import ABC, abstractmethod
from typing import Iterator, List, Union
from pathlib import Path
from time import perf_counter


# EnterpriseDocument and logger are assumed to be defined in previous cells
# from ingestion.models import EnterpriseDocument
# from monitoring.logging import logger


class BaseLoader(ABC):

    def __init__(
        self,
        source: Union[str, Path],
        tenant_id: str = "default",
        security_level: str = "internal",
    ):
        self.source = str(source)
        self.tenant_id = tenant_id
        self.security_level = security_level

        # تفعيل الـ Validation فوراً عند إنشاء الكائن لمنع أي خطأ مبكر
        self.validate_source()

    # =====================================================
    # Context Manager
    # =====================================================

    def __enter__(self):
        logger.info(
            "loader_started",
            extra={
                "loader": self.__class__.__name__,
                "source": self.source,
                "tenant_id": self.tenant_id,
            },
        )
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        if exc_val:
            logger.exception(
                "loader_failed",
                extra={
                    "loader": self.__class__.__name__,
                    "source": self.source,
                },
            )
        else:
            logger.info(
                "loader_completed_successfully",
                extra={"loader": self.__class__.__name__},
            )
        return False  # السماح للـ Exceptions بالارتفاع بعد تسجيلها

    # =====================================================
    # Public API (Template Methods - Do Not Override)
    # =====================================================

    def load(self) -> List[EnterpriseDocument]:
        """
        الأسلوب العام لتحميل المستندات في دفعة واحدة.
        يقوم تلقائياً بالتحقق، الحساب الزمني، وحقن البيانات الأمنية.
        """
        start = perf_counter()

        # استدعاء البنية التحتية الخاصة بالمطور
        raw_docs = self._execute_load()

        # حقن البيانات الأمنية إجبارياً لضمان الـ Compliance
        enriched_docs = [self._enrich_document(doc) for doc in raw_docs]

        elapsed = (perf_counter() - start) * 1000
        logger.info(
            "batch_load_complete",
            extra={
                "loader": self.__class__.__name__,
                "documents_count": len(enriched_docs),
                "latency_ms": round(elapsed, 2),
            },
        )
        return enriched_docs

    def lazy_load(self) -> Iterator[EnterpriseDocument]:
        """
        الأسلوب العام لقراءة البيانات كـ Stream مستمر.
        يحسب الأداء لكل وثيقة ويضمن عدم تسريب أي مستند بدون تأمين.
        """
        start = perf_counter()
        count = 0

        # استهلاك الـ Generator الخاص بالمطور وحقنه أوتوماتيكياً
        for doc in self._execute_lazy_load():
            count += 1
            yield self._enrich_document(doc)

        elapsed = (perf_counter() - start) * 1000
        logger.info(
            "stream_load_complete",
            extra={
                "loader": self.__class__.__name__,
                "total_documents_streamed": count,
                "total_latency_ms": round(elapsed, 2),
            },
        )

    # =====================================================
    # Abstract Methods (To Be Implemented By Subclasses)
    # =====================================================

    @abstractmethod
    def _execute_load(self) -> List[EnterpriseDocument]:
        """المطور يكتب هنا كود القراءة الفعلي للملف بالكامل"""
        raise NotImplementedError

    @abstractmethod
    def _execute_lazy_load(self) -> Iterator[EnterpriseDocument]:
        """المطور يكتب هنا كود الـ Generator (Yield) للقراءة السطرية"""
        raise NotImplementedError

    # =====================================================
    # Validation & Security (Enforced Guardrails)
    # =====================================================

    def validate_source(self):
        if not self.source or str(self.source).strip() == "":
            raise ValueError("Source cannot be empty or blank")

    def _enrich_document(self, doc: EnterpriseDocument) -> EnterpriseDocument:
        """حقن بيانات العميل والأمان بشكل صارم قبل خروج المستند من الـ Loader"""
        doc.tenant_id = self.tenant_id
        doc.security_level = self.security_level
        doc.source = self.source
        return doc
````

````python
from __future__ import annotations

import re
import hashlib
from dataclasses import dataclass
from datetime import datetime, timezone
from typing import Dict, Any, Optional, List, Set, Tuple

try:
    import langdetect
except ImportError:
    langdetect = None

# Removed invalid import: monitoring.logging is not a package in this environment.
# The 'logger' object is already defined in the global scope from cell E71srwBsFkoA.
# from monitoring.logging import logger


# ==========================================================
# Config
# ==========================================================

@dataclass
class ExtractionConfig:
    enable_language_detection: bool = True
    enable_department_detection: bool = True
    enable_pii_flags: bool = True
    min_text_length_for_detection: int = 20
    enable_confidence_score: bool = True
    enable_caching: bool = False


# ==========================================================
# Metadata Extractor (Enterprise Grade)
# ==========================================================

class MetadataExtractor:

    def __init__(self, config: ExtractionConfig = ExtractionConfig()):
        self.config = config

        raw_department_patterns = {
            "legal": r"\b(law|attorney|court|contract|compliance|policy|agreement)\b",
            "finance": r"\b(revenue|invoice|budget|tax|profit|loss|audit|payment)\b",
            "engineering": r"\b(api|code|system|architecture|microservice|kubernetes|database)\b",
            "hr": r"\b(hiring|employee|salary|benefits|recruitment|talent)\b",
            "sales": r"\b(deal|client|pipeline|revenue|crm|marketing)\b",
        }

        self.department_patterns = {
            k: re.compile(v, re.IGNORECASE)
            for k, v in raw_department_patterns.items()
        }

        self.pii_patterns = {
            "email": re.compile(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"),
            "phone": re.compile(r"\+?\d[\d\s\-]{9,}"),
            "id_like": re.compile(r"\b[A-Z0-9]{8,}\b"),
        }

        self.PROTECTED_FIELDS: Set[str] = {
            "tenant_id", "security_level", "source", "processed_at"
        }

        # Simple in-memory cache (enterprise upgrade can be Redis)
        self._cache: Dict[str, Dict[str, Any]] = {}

    # ==========================================================
    # Public API
    # ==========================================================

    def extract(
        self,
        text: Optional[str],
        source: str,
        existing_meta: Optional[Dict[str, Any]] = None,
        tenant_id: str = "default",
        security_level: str = "internal",
    ) -> Dict[str, Any]:

        safe_text = (text or "").strip()
        text_hash = hashlib.sha256(safe_text.encode()).hexdigest()

        # ======================================================
        # Cache Layer (optional)
        # ======================================================
        if self.config.enable_caching and text_hash in self._cache:
            return self._cache[text_hash]

        start_time = datetime.now(timezone.utc)

        base_meta: Dict[str, Any] = {
            "source": source,
            "tenant_id": tenant_id,
            "security_level": security_level,
            "processed_at": start_time.isoformat(),
            "char_count": len(safe_text),
            "word_count": len(safe_text.split()) if safe_text else 0,
        }

        # ======================================================
        # Enrichment
        # ======================================================
        if len(safe_text) >= self.config.min_text_length_for_detection:

            if self.config.enable_language_detection:
                lang = self._detect_language(safe_text)
                base_meta["language"] = lang

            if self.config.enable_department_detection:
                dept, confidence = self._guess_department(safe_text)
                base_meta["department"] = dept
                base_meta["department_confidence"] = confidence

            if self.config.enable_pii_flags:
                base_meta["pii_flags"] = self._detect_pii(safe_text)

        else:
            base_meta.update({
                "language": "unknown",
                "department": "general",
                "pii_flags": []
            })

        # ======================================================
        # Safe Merge
        # ======================================================
        if existing_meta:
            clean = {
                k: v for k, v in existing_meta.items()
                if k not in self.PROTECTED_FIELDS
            }
            base_meta.update(clean)

        # ======================================================
        # Logging
        # ======================================================
        logger.info(
            "metadata_extracted",
            extra={
                "source": source,
                "tenant_id": tenant_id,
                "department": base_meta.get("department"),
                "language": base_meta.get("language"),
                "pii": bool(base_meta.get("pii_flags")),
            },
        )

        # Cache store
        if self.config.enable_caching:
            self._cache[text_hash] = base_meta

        return base_meta

    # ==========================================================
    # Language
    # ==========================================================

    def _detect_language(self, text: str) -> str:
        if not langdetect:
            return "unknown"
        try:
            return langdetect.detect(text)
        except Exception:
            return "unknown"

    # ==========================================================
    # Department (with confidence scoring)
    # ==========================================================

    def _guess_department(self, text: str) -> Tuple[str, float]:

        scores: Dict[str, int] = {}

        for dept, pattern in self.department_patterns.items():
            matches = pattern.findall(text)
            scores[dept] = len(matches)

        if not scores or max(scores.values()) == 0:
            return "general", 0.0

        best = max(scores, key=scores.get)
        total = sum(scores.values())

        confidence = scores[best] / total if total else 0.0

        return best, round(confidence, 3)

    # ==========================================================
    # PII Detection
    # ==========================================================

    def _detect_pii(self, text: str) -> List[str]:
        return [
            k for k, v in self.pii_patterns.items()
            if v.search(text)
        ]
````

````python
from __future__ import annotations

from abc import ABC, abstractmethod
from dataclasses import dataclass
from datetime import datetime, timezone
from typing import Iterator, Iterable, Optional

# لو تعمل في VSCode استخدم:
# from ingestion.models import EnterpriseDocument
# from ingestion.processors.metadata_extractor import MetadataExtractor

# في Colab احذف السطرين السابقين


# ==========================================================
# Hash Store Interface
# ==========================================================

class HashStore(ABC):

    @abstractmethod
    def exists(self, content_hash: str) -> bool:
        pass

    @abstractmethod
    def add(self, content_hash: str) -> None:
        pass


# ==========================================================
# Memory Hash Store
# ==========================================================

class MemoryHashStore(HashStore):

    def __init__(self):
        self._hashes: set[str] = set()

    def exists(self, content_hash: str) -> bool:
        return content_hash in self._hashes

    def add(self, content_hash: str) -> None:
        self._hashes.add(content_hash)


# ==========================================================
# Config
# ==========================================================

@dataclass
class PipelineConfig:

    enable_metadata_extraction: bool = True

    enable_deduplication: bool = True

    reject_empty_documents: bool = True

    minimum_content_length: int = 20

    fail_fast: bool = False


# ==========================================================
# Document Pipeline
# ==========================================================

class DocumentPipeline:

    def __init__(
        self,
        metadata_extractor: Optional["MetadataExtractor"] = None,
        config: Optional[PipelineConfig] = None,
        hash_store: Optional[HashStore] = None,
    ):

        self.metadata_extractor = (
            metadata_extractor
            or MetadataExtractor()
        )

        self.config = (
            config
            or PipelineConfig()
        )

        self.hash_store = (
            hash_store
            or MemoryHashStore()
        )

    # ======================================================
    # Public API
    # ======================================================

    def process(
        self,
        documents: Iterable["EnterpriseDocument"]
    ) -> Iterator["EnterpriseDocument"]:

        processed_count = 0
        skipped_count = 0

        for doc in documents:

            try:

                if not self._validate_document(doc):
                    skipped_count += 1
                    continue

                if not self._deduplicate(doc):
                    skipped_count += 1
                    continue

                self._enrich_metadata(doc)

                # تحديث آخر تعديل بعد المعالجة
                if hasattr(doc, "updated_at"):
                    doc.updated_at = (
                        datetime.now(timezone.utc)
                    )

                processed_count += 1

                yield doc

            except Exception as e:

                logger.error(
                    "document_processing_failed",
                    extra={
                        "document_id": getattr(
                            doc,
                            "id",
                            "unknown"
                        ),
                        "error": str(e),
                    },
                    exc_info=True,
                )

                if self.config.fail_fast:
                    raise

                skipped_count += 1

        logger.info(
            "pipeline_completed",
            extra={
                "processed_docs": processed_count,
                "skipped_docs": skipped_count,
            },
        )

    # ======================================================
    # Validation
    # ======================================================

    def _validate_document(
        self,
        doc: "EnterpriseDocument",
    ) -> bool:

        if (
            self.config.reject_empty_documents
            and not doc.content.strip()
        ):
            logger.warning(
                "empty_document_rejected",
                extra={
                    "document_id": doc.id
                },
            )
            return False

        if (
            len(doc.content)
            < self.config.minimum_content_length
        ):
            logger.warning(
                "short_document_rejected",
                extra={
                    "document_id": doc.id,
                    "length": len(doc.content),
                },
            )
            return False

        return True

    # ======================================================
    # Deduplication
    # ======================================================

    def _deduplicate(
        self,
        doc: "EnterpriseDocument",
    ) -> bool:

        if not self.config.enable_deduplication:
            return True

        if doc.content_hash is None:
            doc.generate_hash()

        if self.hash_store.exists(
            doc.content_hash
        ):
            logger.warning(
                "duplicate_document_skipped",
                extra={
                    "document_id": doc.id
                },
            )
            return False

        self.hash_store.add(
            doc.content_hash
        )

        return True

    # ======================================================
    # Metadata
    # ======================================================

    def _enrich_metadata(
        self,
        doc: "EnterpriseDocument",
    ) -> None:

        if (
            not self.config
            .enable_metadata_extraction
        ):
            return

        extracted = (
            self.metadata_extractor.extract(
                text=doc.content,
                source=doc.source,
                existing_meta=doc.metadata,
                tenant_id=doc.tenant_id,
                security_level=doc.security_level,
            )
        )

        doc.metadata.update(extracted)
````

````python
from __future__ import annotations

import logging
import re
from dataclasses import dataclass
from typing import List, Optional
from uuid import uuid4

try:
    import tiktoken
except ImportError:
    tiktoken = None


# from ingestion.models import EnterpriseDocument, DocumentChunk

logger = logging.getLogger(__name__)


# ==========================================================
# CONFIG
# ==========================================================

@dataclass
class ChunkingConfig:
    chunk_size: int = 512
    chunk_overlap: int = 64

    parent_chunk_size: int = 2048
    parent_chunk_overlap: int = 256

    min_chunk_length_tokens: int = 15

    enable_parent_child: bool = True

    encoding_name: str = "cl100k_base"


# ==========================================================
# SEMANTIC CHUNKER (PRODUCTION GRADE)
# ==========================================================

class SemanticChunker:

    def __init__(self, config: Optional[ChunkingConfig] = None):
        self.config = config or ChunkingConfig()

        if tiktoken:
            self.tokenizer = tiktoken.get_encoding(self.config.encoding_name)
        else:
            self.tokenizer = None
            logger.warning("tiktoken not available → fallback mode")

    # ======================================================
    # PUBLIC API
    # ======================================================

    def chunk_document(self, document) -> List:
        text = (document.content or "").strip()

        if not text:
            return []

        parent_chunks = self._split_by_tokens(
            text,
            self.config.parent_chunk_size,
            self.config.parent_chunk_overlap
        )

        results = []

        for parent_idx, parent in enumerate(parent_chunks):

            parent_id = str(uuid4())

            child_chunks = (
                self._split_by_tokens(
                    parent,
                    self.config.chunk_size,
                    self.config.chunk_overlap
                )
                if self.config.enable_parent_child
                else [parent]
            )

            for child_idx, chunk_text in enumerate(child_chunks):

                token_count = self._count_tokens(chunk_text)

                if token_count < self.config.min_chunk_length_tokens:
                    continue

                chunk = DocumentChunk(
                    document_id=document.id,
                    tenant_id=document.tenant_id,
                    content=chunk_text.strip(),
                    chunk_index=child_idx,
                    parent_chunk_id=parent_id if self.config.enable_parent_child else None,
                    security_level=document.security_level,
                    metadata={
                        **document.metadata,
                        "parent_index": parent_idx,
                    },
                )

                chunk.token_count = token_count
                results.append(chunk)

        logger.info(
            "document_chunked",
            extra={
                "document_id": document.id,
                "chunks": len(results),
            },
        )

        return results

    # ======================================================
    # TOKEN SPLITTING (CLEAN + SAFE)
    # ======================================================

    def _split_by_tokens(self, text: str, max_tokens: int, overlap: int) -> List[str]:

        # تنظيف بسيط قبل التقسيم
        text = re.sub(r'\s+', ' ', text).strip()

        sentences = re.split(r'(?<=[.!?])\s+', text)

        chunks = []
        current = []
        current_tokens = 0

        for sentence in sentences:

            sentence_tokens = self._count_tokens(sentence)

            if sentence_tokens > max_tokens:
                # fallback hard split
                words = sentence.split()

                buffer = []
                buf_tokens = 0

                for w in words:
                    wt = self._count_tokens(w)

                    if buf_tokens + wt > max_tokens:
                        chunks.append(" ".join(buffer))
                        buffer = buffer[-overlap:] if overlap else []
                        buf_tokens = self._count_tokens(" ".join(buffer))

                    buffer.append(w)
                    buf_tokens += wt

                if buffer:
                    current.append(" ".join(buffer))
                    current_tokens += buf_tokens
                continue

            if current_tokens + sentence_tokens > max_tokens:
                chunks.append(" ".join(current))

                # overlap stable (آخر sentences فقط)
                current = current[-2:] if overlap else []
                current_tokens = self._count_tokens(" ".join(current))

            current.append(sentence)
            current_tokens += sentence_tokens

        if current:
            chunks.append(" ".join(current))

        return [c.strip() for c in chunks if c.strip()]

    # ======================================================
    # TOKEN COUNT
    # ======================================================

    def _count_tokens(self, text: str) -> int:

        if self.tokenizer:
            return len(self.tokenizer.encode(text))

        return max(1, len(text) // 4)
````

````python
from dataclasses import dataclass, field
from typing import List, Optional
from uuid import uuid4
import re
import hashlib
import numpy as np

# optional tokenizer
try:
    import tiktoken
    tokenizer = tiktoken.get_encoding("cl100k_base")
except:
    tokenizer = None


# ==========================================================
# CONFIG
# ==========================================================

@dataclass
class ChunkingConfig:
    chunk_size: int = 500
    chunk_overlap: int = 80

    min_chunk_tokens: int = 20
    max_chunk_tokens: int = 800

    semantic_split_threshold: float = 0.78

    enable_deduplication: bool = True
    enable_quality_filter: bool = True


# ==========================================================
# UTIL FUNCTIONS
# ==========================================================

def count_tokens(text: str) -> int:
    if tokenizer:
        return len(tokenizer.encode(text))
    return max(1, len(text) // 4)


def simple_similarity(a: str, b: str) -> float:
    """Lightweight semantic approximation (no heavy model)."""
    set_a = set(a.lower().split())
    set_b = set(b.lower().split())
    if not set_a or not set_b:
        return 0.0
    return len(set_a & set_b) / len(set_a | set_b)


def clean_text(text: str) -> str:
    text = re.sub(r"\s+", " ", text)
    return text.strip()


# ==========================================================
# CHUNK OBJECT
# ==========================================================

@dataclass
class DocumentChunk:
    id: str = field(default_factory=lambda: str(uuid4()))
    content: str = ""
    chunk_index: int = 0
    parent_id: Optional[str] = None
    token_count: int = 0
    quality_score: float = 0.0
    metadata: dict = field(default_factory=dict)
    embedding: Optional[List[float]] = None


# ==========================================================
# ENTERPRISE CHUNKER V2
# ==========================================================

class EnterpriseSemanticChunkerV2:

    def __init__(self, config: ChunkingConfig):
        self.config = config
        self.seen_hashes = set()

    # ----------------------------
    # QUALITY SCORING
    # ----------------------------
    def _score_chunk(self, text: str) -> float:
        words = text.lower().split()

        if not words:
            return 0.0

        uniqueness = len(set(words)) / len(words)
        length_score = min(len(words) / 200, 1.0)

        return (uniqueness * 0.6) + (length_score * 0.4)

    # ----------------------------
    # DEDUP CHECK
    # ----------------------------
    def _is_duplicate(self, text: str) -> bool:
        h = hashlib.md5(text.encode()).hexdigest()

        if h in self.seen_hashes:
            return True

        self.seen_hashes.add(h)
        return False

    # ----------------------------
    # MAIN SPLITTER
    # ----------------------------
    def chunk_document(self, document) -> List[DocumentChunk]:

        text = clean_text(document.content)
        sentences = re.split(r'(?<=[.!?])\s+', text)

        chunks = []
        current = []
        current_tokens = 0
        idx = 0

        for sent in sentences:

            sent_tokens = count_tokens(sent)

            # semantic boundary check (light)
            if current:
                sim = simple_similarity(current[-1], sent)

                if sim < self.config.semantic_split_threshold:
                    self._flush_chunk(
                        chunks,
                        current,
                        idx,
                        document
                    )
                    idx += 1
                    current = []
                    current_tokens = 0

            # overflow check
            if current_tokens + sent_tokens > self.config.chunk_size:
                self._flush_chunk(
                    chunks,
                    current,
                    idx,
                    document
                )
                idx += 1
                current = []
                current_tokens = 0

            current.append(sent)
            current_tokens += sent_tokens

        # flush last
        if current:
            self._flush_chunk(
                chunks,
                current,
                idx,
                document
            )

        return chunks

    # ----------------------------
    # FLUSH LOGIC
    # ----------------------------
    def _flush_chunk(self, chunks, sentences, idx, document):

        text = clean_text(" ".join(sentences))

        if not text:
            return

        token_count = count_tokens(text)

        # filter small chunks
        if token_count < self.config.min_chunk_tokens:
            return

        # deduplication
        if self.config.enable_deduplication and self._is_duplicate(text):
            return

        quality = self._score_chunk(text)

        # quality filter
        if self.config.enable_quality_filter and quality < 0.25:
            return

        chunk = DocumentChunk(
            content=text,
            chunk_index=idx,
            token_count=token_count,
            quality_score=quality,
            metadata={
                "doc_id": document.id,
                "source": document.source,
                "tenant_id": getattr(document, "tenant_id", "default")
            }
        )

        chunks.append(chunk)


# ==========================================================
# TEST RUN
# ==========================================================

doc = EnterpriseDocument(
    content="""
    Artificial Intelligence is transforming enterprise systems.
    Large Language Models enable retrieval augmented generation.
    Vector databases improve search.
    Machine learning improves decision making.
    """,
    source="test.pdf"
)
chunker = EnterpriseSemanticChunkerV2(
    ChunkingConfig(
        chunk_size=120,
        chunk_overlap=30,

        semantic_split_threshold=0.55,   # ↓ مهم جدًا
        min_chunk_tokens=5,              # ↓ مهم جدًا
        enable_quality_filter=False      # 🔥 disable مؤقتًا
    )
)

chunks = chunker.chunk_document(doc)

print("Total Chunks:", len(chunks))

for c in chunks:
    print("\n---")
    print(c.content)
    print("tokens:", c.token_count)
    print("quality:", round(c.quality_score, 3))
````

````python
from dataclasses import dataclass, field
from typing import List
import re
import hashlib
import numpy as np


# ==========================================================
# CONFIG
# ==========================================================

@dataclass
class ContextChunkingConfig:
    min_chunk_tokens: int = 20
    max_chunk_tokens: int = 180

    similarity_threshold: float = 0.45   # lower = more merging
    max_sentences_per_chunk: int = 6


# ==========================================================
# CORE CHUNKER
# ==========================================================

class ContextualChunkerV2:

    def __init__(self, config: ContextChunkingConfig):
        self.config = config

    # --------------------------
    # SIMPLE SIMILARITY
    # --------------------------
    def _similarity(self, a: str, b: str) -> float:
        sa = set(a.lower().split())
        sb = set(b.lower().split())
        if not sa or not sb:
            return 0.0
        return len(sa & sb) / len(sa | sb)

    # --------------------------
    # TOKEN COUNT
    # --------------------------
    def _tokens(self, text: str) -> int:
        return max(1, len(text.split()))

    # --------------------------
    # MAIN FUNCTION
    # --------------------------
    def chunk_document(self, document) -> List:

        text = re.sub(r"\s+", " ", document.content).strip()
        sentences = re.split(r'(?<=[.!?])\s+', text)

        chunks = []
        current = []

        for sent in sentences:

            if not sent:
                continue

            # first sentence
            if not current:
                current.append(sent)
                continue

            last = current[-1]
            sim = self._similarity(last, sent)

            projected = self._tokens(" ".join(current + [sent]))

            # DECISION LOGIC
            if (
                sim >= self.config.similarity_threshold
                and projected <= self.config.max_chunk_tokens
                and len(current) < self.config.max_sentences_per_chunk
            ):
                current.append(sent)
            else:
                chunks.append(" ".join(current))
                current = [sent]

        if current:
            chunks.append(" ".join(current))

        # FILTER SMALL CHUNKS
        final_chunks = [
            c for c in chunks
            if self._tokens(c) >= self.config.min_chunk_tokens
        ]

        return final_chunks
````

````python
import numpy as np
from dataclasses import dataclass
from typing import List
import re


# ==========================================================
# CONFIG
# ==========================================================

@dataclass
class ContextChunkingConfig:
    similarity_threshold: float = 0.55
    max_chunk_tokens: int = 180
    min_chunk_tokens: int = 10
    max_sentences_per_chunk: int = 6


# ==========================================================
# ENTERPRISE CHUNKER (OPTIMIZED)
# ==========================================================

class ContextualChunkerOptimized:

    def __init__(self, config, embedder):
        self.config = config
        self.embedder = embedder  # ⌁ injected once (NO repeated calls)

    # ------------------------------------------------------
    def _tokens(self, text: str) -> int:
        return max(1, len(text.split()))

    # ------------------------------------------------------
    def _cosine(self, a: np.ndarray, b: np.ndarray) -> float:
        return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))

    # ------------------------------------------------------
    def chunk_document(self, document) -> List[str]:

        text = re.sub(r"\s+", " ", document.content).strip()
        sentences = re.split(r'(?<=[.!?])\s+', text)

        if not sentences:
            return []

        # ==================================================
        # 1. PRECOMPUTE EMBEDDINGS (⌁ MAIN OPTIMIZATION)
        # ==================================================
        embeddings = self.embedder.encode(sentences, normalize_embeddings=True)

        chunks = []
        current_chunk = []
        current_embs = []

        # running centroid for fast similarity
        centroid = None

        for i, sent in enumerate(sentences):

            sent_emb = embeddings[i]

            if not current_chunk:
                current_chunk.append(sent)
                current_embs.append(sent_emb)
                centroid = sent_emb
                continue

            # update centroid
            centroid = np.mean(current_embs, axis=0)
            sim_centroid = self._cosine(centroid, sent_emb)
            sim_last = self._cosine(current_embs[-1], sent_emb)

            sim = (0.5 * sim_centroid) + (0.5 * sim_last)

            # – NEW: topic shift detection
            topic_shift = sim_last < 0.45

            projected_tokens = self._tokens(" ".join(current_chunk + [sent]))

            # ==================================================
            # DECISION
            # ==================================================
            if (
                not topic_shift
                and sim >= self.config.similarity_threshold
                and projected_tokens <= self.config.max_chunk_tokens
            ):
                current_chunk.append(sent)
                current_embs.append(sent_emb)
            else:
                chunks.append(" ".join(current_chunk))
                current_chunk = [sent]
                current_embs = [sent_emb]
                centroid = sent_emb

        if current_chunk:
            chunks.append(" ".join(current_chunk))

        # ==================================================
        # FILTER SMALL CHUNKS
        # ==================================================
        chunks = [
            c for c in chunks
            if self._tokens(c) >= self.config.min_chunk_tokens
        ]

        return chunks
````

````python
from __future__ import annotations

from abc import ABC, abstractmethod
from dataclasses import dataclass
from hashlib import sha256
from math import sqrt
from time import perf_counter, sleep
from typing import Any, Dict, Iterable, List, Optional, Sequence

# Optional dependencies
try:
    from sentence_transformers import SentenceTransformer
except ImportError:
    SentenceTransformer = None

try:
    from openai import OpenAI
except ImportError:
    OpenAI = None

try:
    import requests
except ImportError:
    requests = None


# ==========================================================
# CONFIG
# ==========================================================

@dataclass
class EmbeddingConfig:
    batch_size: int = 32
    max_retries: int = 3
    base_backoff_seconds: float = 0.6
    normalize_embeddings: bool = True
    enable_cache: bool = True
    embedding_version: str = "v1"
    expected_dimension: Optional[int] = None


# ==========================================================
# UTILS
# ==========================================================

def _l2_normalize(vec: Sequence[float]) -> List[float]:
    norm = sqrt(sum(x * x for x in vec))
    if norm == 0:
        return list(vec)
    return [x / norm for x in vec]


def _batch(items: Sequence[Any], batch_size: int) -> Iterable[List[Any]]:
    for i in range(0, len(items), batch_size):
        yield list(items[i:i + batch_size])


def _text_hash(provider_name: str, model_name: str, text: str) -> str:
    payload = f"{provider_name}::{model_name}::{text}".encode("utf-8")
    return sha256(payload).hexdigest()


# ==========================================================
# PROVIDER INTERFACE
# ==========================================================

class EmbeddingProvider(ABC):
    @property
    @abstractmethod
    def provider_name(self) -> str:
        raise NotImplementedError

    @property
    @abstractmethod
    def model_name(self) -> str:
        raise NotImplementedError

    @abstractmethod
    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        raise NotImplementedError


# ==========================================================
# LOCAL PROVIDER
# ==========================================================

class LocalEmbeddingProvider(EmbeddingProvider):
    """
    Production-friendly local provider using sentence-transformers.
    """

    def __init__(
        self,
        model_name: str = "BAAI/bge-small-en-v1.5",
        device: Optional[str] = None,
        normalize_embeddings: bool = True,
    ):
        if SentenceTransformer is None:
            raise ImportError(
                "sentence-transformers is not installed. "
                "Run: pip install sentence-transformers"
            )

        self._model_name = model_name
        self._normalize_embeddings = normalize_embeddings
        self.model = SentenceTransformer(model_name, device=device)

    @property
    def provider_name(self) -> str:
        return "local"

    @property
    def model_name(self) -> str:
        return self._model_name

    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        embeddings = self.model.encode(
            texts,
            normalize_embeddings=self._normalize_embeddings,
            show_progress_bar=False,
        )
        return embeddings.tolist()


# ==========================================================
# OPENAI PROVIDER
# ==========================================================

class OpenAIEmbeddingProvider(EmbeddingProvider):
    def __init__(
        self,
        api_key: str,
        model_name: str = "text-embedding-3-small",
    ):
        if OpenAI is None:
            raise ImportError("openai package is not installed. Run: pip install openai")

        self.client = OpenAI(api_key=api_key)
        self._model_name = model_name

    @property
    def provider_name(self) -> str:
        return "openai"

    @property
    def model_name(self) -> str:
        return self._model_name

    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        response = self.client.embeddings.create(
            model=self._model_name,
            input=texts,
        )
        return [item.embedding for item in response.data]


# ==========================================================
# JINA PROVIDER
# ==========================================================

class JinaEmbeddingProvider(EmbeddingProvider):
    def __init__(
        self,
        api_key: str,
        model_name: str = "jina-embeddings-v3",
    ):
        if requests is None:
            raise ImportError("requests package is not installed. Run: pip install requests")

        self.api_key = api_key
        self._model_name = model_name

    @property
    def provider_name(self) -> str:
        return "jina"

    @property
    def model_name(self) -> str:
        return self._model_name

    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        response = requests.post(
            "https://api.jina.ai/v1/embeddings",
            headers={"Authorization": f"Bearer {self.api_key}"},
            json={"model": self._model_name, "input": texts},
            timeout=60,
        )
        response.raise_for_status()
        payload = response.json()
        return [item["embedding"] for item in payload["data"]]


# ==========================================================
# MANAGER
# ==========================================================

class EmbeddingManager:
    """
    Responsible for:
    - batching
    - retries
    - caching
    - normalization
    - attaching embeddings + metadata back to chunks
    """

    def __init__(
        self,
        provider: EmbeddingProvider,
        config: Optional[EmbeddingConfig] = None,
    ):
        self.provider = provider
        self.config = config or EmbeddingConfig()
        self.cache: Dict[str, List[float]] = {}

    def _embed_batch_with_retry(self, texts: List[str]) -> List[List[float]]:
        last_exc = None

        for attempt in range(1, self.config.max_retries + 1):
            try:
                return self.provider.embed_documents(texts)
            except Exception as exc:
                last_exc = exc
                if attempt == self.config.max_retries:
                    break
                sleep(self.config.base_backoff_seconds * attempt)

        raise RuntimeError(
            f"Embedding failed after {self.config.max_retries} retries "
            f"using provider={self.provider.provider_name}, model={self.provider.model_name}"
        ) from last_exc

    def _cache_lookup(self, text: str) -> Optional[List[float]]:
        if not self.config.enable_cache:
            return None
        key = _text_hash(self.provider.provider_name, self.provider.model_name, text)
        return self.cache.get(key)

    def _cache_store(self, text: str, embedding: List[float]) -> None:
        if not self.config.enable_cache:
            return
        key = _text_hash(self.provider.provider_name, self.provider.model_name, text)
        self.cache[key] = embedding

    def embed_texts(self, texts: List[str]) -> List[List[float]]:
        if not texts:
            return []

        final_embeddings: List[Optional[List[float]]] = [None] * len(texts)
        missing_texts: List[str] = []
        missing_positions: List[int] = []

        # 1) cache hit
        for i, text in enumerate(texts):
            cached = self._cache_lookup(text)
            if cached is not None:
                final_embeddings[i] = cached
            else:
                missing_positions.append(i)
                missing_texts.append(text)

        # 2) provider inference in batches
        if missing_texts:
            start = perf_counter()
            generated: List[List[float]] = []

            for batch_texts in _batch(missing_texts, self.config.batch_size):
                batch_embeddings = self._embed_batch_with_retry(batch_texts)
                generated.extend(batch_embeddings)

            latency_ms = (perf_counter() - start) * 1000

            # 3) normalize + cache + place back
            for pos, text, emb in zip(missing_positions, missing_texts, generated):
                if self.config.normalize_embeddings:
                    emb = _l2_normalize(emb)

                if self.config.expected_dimension is not None and len(emb) != self.config.expected_dimension:
                    raise ValueError(
                        f"Embedding dimension mismatch: expected {self.config.expected_dimension}, got {len(emb)}"
                    )

                self._cache_store(text, emb)
                final_embeddings[pos] = emb

            if "logger" in globals():
                logger.info(
                    "embedding_batch_complete",
                    extra={
                        "provider": self.provider.provider_name,
                        "model": self.provider.model_name,
                        "count": len(texts),
                        "cached": len(texts) - len(missing_texts),
                        "generated": len(missing_texts),
                        "latency_ms": round(latency_ms, 2),
                    },
                )

        return [emb for emb in final_embeddings if emb is not None]

    def embed_chunks(self, chunks: List[Any]) -> List[Any]:
        if not chunks:
            return []

        texts = [getattr(chunk, "content", "") or "" for chunk in chunks]

        # Safety: skip empty strings rather than crashing provider calls
        non_empty_positions = [i for i, t in enumerate(texts) if t.strip()]
        non_empty_texts = [texts[i] for i in non_empty_positions]

        if not non_empty_texts:
            return chunks

        embeddings = self.embed_texts(non_empty_texts)

        for idx, emb in zip(non_empty_positions, embeddings):
            chunk = chunks[idx]
            chunk.embedding = emb

            # Attach rich metadata if the chunk supports it
            if not hasattr(chunk, "embedding_metadata") or chunk.embedding_metadata is None:
                try:
                    chunk.embedding_metadata = {}
                except Exception:
                    pass

            meta = {
                "provider": self.provider.provider_name,
                "model_name": self.provider.model_name,
                "embedding_version": self.config.embedding_version,
                "dimension": len(emb),
                "normalized": self.config.normalize_embeddings,
            }

            if hasattr(chunk, "embedding_metadata"):
                try:
                    chunk.embedding_metadata.update(meta)
                except Exception:
                    pass

            if hasattr(chunk, "metadata") and isinstance(chunk.metadata, dict):
                chunk.metadata.update({
                    "embedding_provider": self.provider.provider_name,
                    "embedding_model": self.provider.model_name,
                    "embedding_version": self.config.embedding_version,
                    "embedding_dimension": len(emb),
                })

        return chunks


# ==========================================================
# BUILDER
# ==========================================================

def build_embedding_provider(
    provider_type: str = "local",
    *,
    local_model_name: str = "BAAI/bge-small-en-v1.5",
    openai_api_key: Optional[str] = None,
    openai_model_name: str = "text-embedding-3-small",
    jina_api_key: Optional[str] = None,
    jina_model_name: str = "jina-embeddings-v3",
    normalize_embeddings: bool = True,
) -> EmbeddingProvider:
    provider_type = provider_type.lower().strip()

    if provider_type == "local":
        return LocalEmbeddingProvider(
            model_name=local_model_name,
            normalize_embeddings=normalize_embeddings,
        )

    if provider_type == "openai":
        if not openai_api_key:
            raise ValueError("openai_api_key is required for provider_type='openai'")
        return OpenAIEmbeddingProvider(
            api_key=openai_api_key,
            model_name=openai_model_name,
        )

    if provider_type == "jina":
        if not jina_api_key:
            raise ValueError("jina_api_key is required for provider_type='jina'")
        return JinaEmbeddingProvider(
            api_key=jina_api_key,
            model_name=jina_model_name,
        )

    raise ValueError(f"Unsupported provider_type: {provider_type}")


# ==========================================================
# EXECUTION EXAMPLE
# ==========================================================

# Use your already-created chunks from the previous stage
# If chunks are not available, create a tiny fallback demo.
if "chunks" not in globals():
    print("⚠️ chunks not found. Creating a demo document from scratch...")

    try:
        demo_doc = EnterpriseDocument(
            content="Enterprise RAG systems need stable embeddings. " * 30,
            source="demo.pdf",
            tenant_id="tenant_demo"
        )
        demo_chunker = SemanticChunker(ChunkingConfig(chunk_size=120))
        chunks = demo_chunker.chunk_document(demo_doc)
    except Exception:
        class DemoChunk:
            def __init__(self, content):
                self.content = content
                self.embedding = None
                self.metadata = {}

        chunks = [
            DemoChunk("Enterprise RAG systems need stable embeddings."),
            DemoChunk("Vector representations must be normalized and cached."),
        ]

provider = build_embedding_provider(
    provider_type="local",
    local_model_name="BAAI/bge-small-en-v1.5",
    normalize_embeddings=True,
)

embedding_manager = EmbeddingManager(
    provider=provider,
    config=EmbeddingConfig(
        batch_size=16,
        max_retries=3,
        base_backoff_seconds=0.5,
        normalize_embeddings=True,
        enable_cache=True,
        embedding_version="v10",
        expected_dimension=None,   # optional, set if you want strict validation
    )
)

embedded_chunks = embedding_manager.embed_chunks(chunks)

print(f"✅ Embedded chunks: {len(embedded_chunks)}")
if embedded_chunks:
    print(f"Provider: {provider.provider_name}")
    print(f"Model: {provider.model_name}")
    print(f"Vector dim: {len(embedded_chunks[0].embedding)}")
    print(f"Sample metadata: {getattr(embedded_chunks[0], 'metadata', {})}")
````

````python
import time
import hashlib
import numpy as np
from collections import Counter

# =====================================================
# STAGE 10.5 — ADVANCED ENTERPRISE EMBEDDING PIPELINE
# =====================================================

class AdvancedEmbeddingEngine:
    def __init__(self, embedding_manager):
        self.embedding_manager = embedding_manager
        self.stats = {}

    # --------------------------------------------
    # Dynamic Batch Size
    # --------------------------------------------
    def choose_batch_size(self, n_chunks):
        if n_chunks < 50:
            return 8
        elif n_chunks < 500:
            return 16
        elif n_chunks < 5000:
            return 32
        return 64

    # --------------------------------------------
    # Semantic Enrichment
    # --------------------------------------------
    def enrich_chunk(self, chunk):
        content = getattr(chunk, "content", "")

        metadata = getattr(chunk, "metadata", {}) or {}

        source = metadata.get("source", "unknown_document")
        section = metadata.get("section", "general")

        enriched = f"""
Document: {source}
Section: {section}
Content:
{content}
"""
        return enriched.strip()

    # --------------------------------------------
    # Duplicate Detection
    # --------------------------------------------
    def detect_duplicates(self, chunks):
        hashes = []
        duplicates = 0

        for chunk in chunks:
            text = getattr(chunk, "content", "")
            h = hashlib.sha256(text.encode()).hexdigest()
            hashes.append(h)

        counts = Counter(hashes)

        for _, c in counts.items():
            if c > 1:
                duplicates += c - 1

        duplicate_ratio = duplicates / len(chunks) if chunks else 0
        return duplicates, duplicate_ratio

    # --------------------------------------------
    # Validation
    # --------------------------------------------
    def validate_embedding(self, vec):
        vec = np.array(vec)

        if np.isnan(vec).any():
            raise ValueError("Embedding contains NaN")

        if np.isinf(vec).any():
            raise ValueError("Embedding contains Inf")

        if np.linalg.norm(vec) == 0:
            raise ValueError("Zero vector detected")

        return True

    # --------------------------------------------
    # Quality Metrics
    # --------------------------------------------
    def compute_quality_metrics(self, chunks, elapsed):
        vectors = [
            np.array(chunk.embedding)
            for chunk in chunks
            if hasattr(chunk, "embedding") and chunk.embedding is not None
        ]

        norms = [np.linalg.norm(v) for v in vectors]

        metrics = {
            "embedded_chunks": len(vectors),
            "latency_sec": round(elapsed, 3),
            "throughput_chunks_per_sec": round(len(vectors) / elapsed, 2) if elapsed else 0,
            "avg_vector_norm": round(float(np.mean(norms)), 4) if norms else 0
        }

        return metrics

    # --------------------------------------------
    # Main Pipeline
    # --------------------------------------------
    def process(self, chunks):
        if not chunks:
            raise ValueError("No chunks found. Run chunking stage first.")

        duplicates, dup_ratio = self.detect_duplicates(chunks)

        batch_size = self.choose_batch_size(len(chunks))

        if hasattr(self.embedding_manager, "config"):
            self.embedding_manager.config.batch_size = batch_size

        enriched_texts = []

        for chunk in chunks:
            enriched = self.enrich_chunk(chunk)
            enriched_texts.append(enriched)

        start = time.time()

        embeddings = self.embedding_manager.embed_texts(enriched_texts)

        for chunk, emb in zip(chunks, embeddings):
            self.validate_embedding(emb)
            chunk.embedding = emb

            if not hasattr(chunk, "metadata"):
                chunk.metadata = {}

            chunk.metadata["advanced_embedding"] = True
            chunk.metadata["validated"] = True

        elapsed = time.time() - start

        quality = self.compute_quality_metrics(chunks, elapsed)

        quality["duplicates_found"] = duplicates
        quality["duplicate_ratio"] = round(dup_ratio, 4)
        quality["dynamic_batch_size"] = batch_size

        self.stats = quality

        return chunks


# =====================================================
# EXECUTION
# =====================================================

if "chunks" not in globals():
    raise RuntimeError("chunks variable not found. Please run chunking stage first.")

advanced_engine = AdvancedEmbeddingEngine(embedding_manager)

embedded_chunks = advanced_engine.process(chunks)

print("\n===== STAGE 10.5 REPORT =====")
for k, v in advanced_engine.stats.items():
    print(f"{k}: {v}")

print("\nSample Metadata:")
print(embedded_chunks[0].metadata)
````

````python
import requests
import json
from dataclasses import dataclass
from typing import List, Dict

# =====================================================
# STAGE 12 — ENTERPRISE RETRIEVAL ORCHESTRATOR
# =====================================================

NGROK_URL = "https://giggly-repayment-sizable.ngrok-free.dev"
QWEN_MODEL = "qwen2.5-coder-7b-instruct"


@dataclass
class RetrievalConfig:
    top_k_dense: int = 5
    top_k_sparse: int = 5
    final_top_k: int = 4
    max_context_chars: int = 4000


class QueryAnalyzer:
    def analyze(self, query: str):
        q = query.lower()

        if any(word in q for word in ["error", "bug", "traceback", "exception"]):
            return "debug"

        if any(word in q for word in ["config", ".env", "setting"]):
            return "config"

        if any(word in q for word in ["code", "python", "function", "class"]):
            return "code"

        return "general"


class RetrievalOrchestrator:
    def __init__(self, hybrid_retriever, config):
        self.hybrid_retriever = hybrid_retriever
        self.config = config
        self.query_analyzer = QueryAnalyzer()

    def rerank(self, dense_results, sparse_results):
        merged = []
        seen = set()

        for r in dense_results:
            key = r["content"][:120]
            if key not in seen:
                merged.append({
                    "score": float(r["score"]) + 5,
                    "content": r["content"]
                })
                seen.add(key)

        for r in sparse_results:
            key = r["content"][:120]
            if key not in seen:
                merged.append({
                    "score": float(r["score"]),
                    "content": r["content"]
                })
                seen.add(key)

        merged.sort(key=lambda x: x["score"], reverse=True)
        return merged[:self.config.final_top_k]

    def compress_context(self, results):
        context = []

        for i, result in enumerate(results):
            text = result["content"]

            if len(text) > 800:
                text = text[:800]

            context.append(f"[Chunk {i+1}]\n{text}")

        final_context = "\n\n".join(context)
        return final_context[:self.config.max_context_chars]

    def retrieve(self, query):
        query_type = self.query_analyzer.analyze(query)

        # This is where hybrid_retriever is used and causes the NameError if not defined.
        results = self.hybrid_retriever.search(query)

        reranked = self.rerank(
            results["dense_results"],
            results["sparse_results"]
        )

        context = self.compress_context(reranked)

        return {
            "query_type": query_type,
            "context": context
        }


class PromptBuilder:
    def build(self, query, context, query_type):
        system_prompt = f"""
You are an enterprise RAG assistant specialized in AI engineering.

Query type: {query_type}

Rules:
1. Answer only from retrieved context if possible
2. If context insufficient, explicitly say so
3. Be precise and technical
4. Avoid hallucinations
"""

        user_prompt = f"""
Retrieved Context:
{context}

User Question:
{query}

Answer:
"""

        return system_prompt, user_prompt


class QwenGenerator:
    def __init__(self, endpoint):
        self.endpoint = endpoint

    def generate(self, system_prompt, user_prompt):
        payload = {
            "model": QWEN_MODEL,
            "messages": [
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt}
            ],
            "temperature": 0.2,
            "max_tokens": 700
        }

        response = requests.post(
            f"{self.endpoint}/v1/chat/completions",
            json=payload,
            timeout=120
        )

        result = response.json()

        return result["choices"][0]["message"]["content"]


# =====================================================
# EXECUTION
# =====================================================

# Check if hybrid_retriever is defined. If not, raise an informative error.
if 'hybrid_retriever' not in globals():
    raise NameError(
        "hybrid_retriever is not defined. Please ensure Stage 11 (the cell defining "
        "EnterpriseVectorStore, SparseRetriever, and HybridRetriever) has been executed successfully."
        " If you have run previous cells, consider re-running them from the start of the notebook." # Added specific guidance to rerun from the start of the notebook.
    )

retrieval_config = RetrievalConfig()

orchestrator = RetrievalOrchestrator(
    hybrid_retriever,
    retrieval_config
)

prompt_builder = PromptBuilder()
generator = QwenGenerator(NGROK_URL)

print("Stage 12 initialized successfully.")
````

````python
from dataclasses import dataclass
from collections import Counter
import math
import hashlib

@dataclass
class ChunkQualityMetrics:
    quality_score: float
    token_count: int
    lexical_diversity: float
    duplicate_ratio: float


class EnterpriseChunkEvaluator:

    def __init__(self):
        self.hashes = set()

    def evaluate(self, chunk_text: str):

        words = chunk_text.lower().split()

        token_count = len(words)

        if token_count == 0:
            return ChunkQualityMetrics(
                0.0, 0, 0.0, 1.0
            )

        lexical_diversity = (
            len(set(words))
            / token_count
        )

        chunk_hash = hashlib.md5(
            chunk_text.encode()
        ).hexdigest()

        duplicate_ratio = (
            1.0
            if chunk_hash in self.hashes
            else 0.0
        )

        self.hashes.add(chunk_hash)

        quality_score = (
            lexical_diversity * 0.5
            + min(token_count / 300, 1.0) * 0.5
        )

        quality_score *= (
            1 - duplicate_ratio
        )

        return ChunkQualityMetrics(
            quality_score,
            token_count,
            lexical_diversity,
            duplicate_ratio
        )


class RetrievalFusionEngine:

    @staticmethod
    def reciprocal_rank_fusion(
        dense_results,
        sparse_results,
        k=60
    ):

        scores = {}

        for rank, item in enumerate(
            dense_results,
            start=1
        ):
            key = item["content"]

            scores[key] = (
                scores.get(key, 0)
                + 1 / (k + rank)
            )

        for rank, item in enumerate(
            sparse_results,
            start=1
        ):
            key = item["content"]

            scores[key] = (
                scores.get(key, 0)
                + 1 / (k + rank)
            )

        ranked = sorted(
            scores.items(),
            key=lambda x: x[1],
            reverse=True
        )

        return ranked


evaluator = EnterpriseChunkEvaluator()

for chunk in embedded_chunks:

    metrics = evaluator.evaluate(
        chunk.content
    )

    chunk.metadata["quality_score"] = (
        metrics.quality_score
    )

    chunk.metadata["lexical_diversity"] = (
        metrics.lexical_diversity
    )

    chunk.metadata["duplicate_ratio"] = (
        metrics.duplicate_ratio
    )

print(
    f"✅ Evaluated {len(embedded_chunks)} chunks"
)

avg_quality = sum(
    c.metadata["quality_score"]
    for c in embedded_chunks
) / len(embedded_chunks)

print(
    f"Average Quality Score: {avg_quality:.3f}"
)
````

````python
from dataclasses import dataclass
from typing import List, Dict
import numpy as np


# ==========================================================
# 1. QUERY ANALYZER
# ==========================================================

class QueryAnalyzer:

    def classify(self, query: str) -> str:

        q = query.lower()

        if any(k in q for k in ["how", "why", "explain"]):
            return "explanatory"

        if any(k in q for k in ["what", "define", "who"]):
            return "factual"

        if any(k in q for k in ["compare", "difference"]):
            return "comparative"

        return "general"


# ==========================================================
# 2. RETRIEVAL ROUTER
# ==========================================================

class RetrievalRouter:

    def get_top_k(self, query_type: str) -> int:

        return {
            "factual": 3,
            "explanatory": 6,
            "comparative": 8,
            "general": 5
        }.get(query_type, 5)


# ==========================================================
# 3. RERANKER
# ==========================================================

class ChunkReranker:

    def rerank(self, results: List[Dict]) -> List[Dict]:

        def score(x):
            return x["score"] - (len(x["content"]) * 0.0001)

        return sorted(results, key=score, reverse=True)


# ==========================================================
# 4. FINAL RAG ORCHESTRATOR
# ==========================================================

class EnterpriseRAG:

    def __init__(self, retriever, embedding_manager):

        self.retriever = retriever
        self.embedding_manager = embedding_manager

        self.analyzer = QueryAnalyzer()
        self.router = RetrievalRouter()
        self.reranker = ChunkReranker()

    def retrieve(self, query: str):

        # 1. classify query
        qtype = self.analyzer.classify(query)

        # 2. decide top_k
        top_k = self.router.get_top_k(qtype)

        # 3. embed query
        query_emb = self.embedding_manager.embed_texts([query])[0]

        # 4. retrieve
        results = self.retriever.vector_store.search(query_emb, top_k)

        # 5. rerank
        results = self.reranker.rerank(results)

        return {
            "query_type": qtype,
            "top_k": top_k,
            "results": results
        }
````

````python
from dataclasses import dataclass
from typing import List, Dict


# ==========================================================
# 1. DOMAIN ROUTER (NEW CORE LAYER)
# ==========================================================

class DomainRouter:

    def route(self, query: str) -> str:

        q = query.lower()

        # SYSTEM / CONFIGURATION INTENT
        if any(k in q for k in [
            "qdrant", "url", "config", "env", "environment",
            "localhost", "docker", "api key"
        ]):
            return "system"

        # CODE / IMPLEMENTATION INTENT
        if any(k in q for k in [
            "function", "class", "code", "implement", "error",
            "exception", "debug"
        ]):
            return "code"

        # AI / ML INTENT
        if any(k in q for k in [
            "llm", "embedding", "vector", "rag", "model",
            "transformer", "ai"
        ]):
            return "ai"

        return "general"


# ==========================================================
# 2. ROUTED RETRIEVAL WRAPPER
# ==========================================================

class RoutedRetriever:

    def __init__(self, vector_store, sparse_retriever, embedding_manager):
        self.vector_store = vector_store
        self.sparse_retriever = sparse_retriever
        self.embedding_manager = embedding_manager
        self.router = DomainRouter()

    def search(self, query: str, top_k: int = 5):

        domain = self.router.route(query)

        query_embedding = self.embedding_manager.embed_texts([query])[0]

        # ===== DOMAIN-AWARE FILTERING =====
        dense_results = self.vector_store.search(
            query_embedding,
            top_k,
            domain=domain
        )

        # FILTER BY DOMAIN (IMPORTANT FIX)
        filtered = []
        for r in dense_results:
            content = r["content"].lower()

            if domain == "system" and any(x in content for x in ["config", "qdrant", "url"]):
                filtered.append(r)

            elif domain == "ai" and any(x in content for x in ["llm", "vector", "embedding", "rag"]):
                filtered.append(r)

            elif domain == "code" and any(x in content for x in ["def", "class", "import"]):
                filtered.append(r)

            elif domain == "general":
                filtered.append(r)

        # fallback if filtering removes everything
        if not filtered:
            filtered = dense_results

        sparse_results = self.sparse_retriever.search(query, top_k)

        return {
            "domain": domain,
            "dense": filtered,
            "sparse": sparse_results
        }
````
